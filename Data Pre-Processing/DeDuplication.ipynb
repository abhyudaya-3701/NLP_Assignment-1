```python
import os
```

```python
file_path = '/mnt/HDFS1/language_nlp/english_nlp_team17/Assignment1/good_files.txt'

file_paths = []

with open(file_path, 'r') as file:
    file_paths = [line.strip() for line in file.readlines()]  
```

```python
import pandas as pd
from datasketch import MinHash
import os

data = {"file_path": [], "content": [], "remove_mask": []}

for file_path in file_paths:
    if os.path.exists(file_path):
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()
            data["file_path"].append(file_path)
            data["content"].append(content)
            data["remove_mask"].append(0) 
    else:
        print(f"File not found: {file_path}")

df = pd.DataFrame(data)

threshold = 0.7  

num_docs = len(df)

for i in range(num_docs):
    content1 = df.iloc[i]["content"]
    m1 = MinHash()  
    for token in content1.split():
        m1.update(token.encode('utf8'))
    
    for j in range(i + 1, num_docs):
        content2 = df.iloc[j]["content"]
        m2 = MinHash()  
        for token in content2.split():
            m2.update(token.encode('utf8'))
        
        similarity = m1.jaccard(m2)
        
        if similarity > threshold:
            df.at[i, "remove_mask"] = 1  
            break  

print(df)

filtered_df = df[df["remove_mask"] == 0]

filtered_df.to_csv("filtered_documents.csv", index=False)

print(filtered_df)
```
