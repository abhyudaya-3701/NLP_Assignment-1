```python
import os
import json


with open("/mnt/HDFS1/language_nlp/english_nlp_team17/Assignment1/Shiva/Scripts/Bad_words.txt", "r") as file:
    badwords_set = set(line.strip().lower() for line in file)


def contains_bad_words(file_path, badwords_set):
    
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            text = f.read().lower()
            for badword in badwords_set:
                if badword in text:
                    return True
    except (IsADirectoryError, UnicodeDecodeError):
        
        # print(f"Skipping file: {file_path}")
        return False
    return False



badword_cleaned_source_dict = {}
badword_src_file_dict = {}

for source, file_list in src_file_dict.items():
    i = 0
    j =0
    print(source[0])
    total_size = 0
    valid_files = []
    badword_src_file_dict[source[0]] = []
    for file in file_list:

        file_path = os.path.join(source[1], file)
        # print(file_path)
        if not file_path.endswith('.txt'):
            # print(f"Skipping non-text file: {file_path}")
            continue
        if i%200 == 0:
            print(f'{source[0]} : {i}')
            i= i + 1
        
        if contains_bad_words(file_path, badwords_set):
            # print(f"File {file_path} contains bad words, skipping.")
            continue 
        badword_src_file_dict[source[0]].append(file)
        print(f'Files with no badwords in {source[0]} : {j}')
        j = j + 1
        with open('badword_src_file_dict_1.txt', 'w') as file:
            json.dump(badword_src_file_dict, file)
        
        total_size += os.path.getsize(file_path)
        valid_files.append(file_path)

    badword_cleaned_source_dict[source[0]] = (total_size / (1024 * 1024 * 1024), len(valid_files))

```

```python

badword_cleaned_source_df = pd.DataFrame(badword_cleaned_source_dict).transpose()

columns = ['Volume(in GB)', 'Number of files']
badword_cleaned_source_df.columns = columns
badword_cleaned_source_df['Number of files'] = badword_cleaned_source_df['Number of files'].round().astype('int')


print(badword_cleaned_source_df)
```
