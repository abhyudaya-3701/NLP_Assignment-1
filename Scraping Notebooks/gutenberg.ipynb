```python
import requests
from bs4 import BeautifulSoup
import os
import time
from requests.exceptions import ChunkedEncodingError, ConnectionError
```

```python
base_url = 'https://www.gutenberg.org'
```

```python
def fetch_books_by_author(letter):
    response = requests.get(f'{base_url}/browse/authors/{letter}')
    soup = BeautifulSoup(response.text, 'html.parser')
    books = []
    for entry in soup.find_all('li', class_='pgdbetext'):
        link = entry.find('a', href=True)
        if link and '(English)' in entry.text:
            books.append(base_url + link['href'])
    print(f'{letter} : {len(books)} books found')
    return books
```

```python
def extract_book_id(url):
    return url.strip('/').split('/')[-1]
```

```python
def save_book(book_url, directory='Data/gutenberg_data'):
    os.makedirs(directory, exist_ok=True)
    book_id = extract_book_id(book_url)
    download_link = f'{base_url}/cache/epub/{book_id}/pg{book_id}.txt'
    
    try:
        response = requests.get(download_link, stream=True)
        if response.status_code == 200:
            file_path = os.path.join(directory, f'{book_id}.txt')
            with open(file_path, 'wb') as file:
                for chunk in response.iter_content(1024):
                    if chunk:
                        file.write(chunk)
        else:
            print(f"Failed to download from {download_link}")
    except:
        print(f"Failure for link {download_link}")
```

```python
for letter in range(ord('n'), ord('z') + 1):
    books = fetch_books_by_author(chr(letter))
    for i, book_url in enumerate(books):
        save_book(book_url)
        if (i + 1) % 10 == 0:
            print(f'{i + 1} books downloaded from {book_url}')
        time.sleep(1)
```
